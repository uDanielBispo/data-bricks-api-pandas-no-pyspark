```
url = 'https://pt.wikipedia.org/wiki/Educa%C3%A7%C3%A3o'
```

```
lista = ps.read_html(url)
```

```
df_edu = lista[1]
df_edu
```

```
df_edu = df_edu.drop(0)
df_edu
```

```
df_edu.columns = df_edu.iloc[0]
df_edu = df_edu.iloc[1:]
df_edu
```

```
df_edu = df_edu.rename(columns={'Idade': 'País'})
df_edu
```

```
df_edu.to_parquet('/FileStore/tables/edu.parquet', partition_cols='País')
```

```
ps.read_parquet('/FileStore/tables/edu.parquet')
```
